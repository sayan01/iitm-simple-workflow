```python
import json
d = json.load(open('iitm-courses.json'))
```

```python
d
```

```python
courses = [x for x in d if x['course_level'] < 3]
```

```python
courses
```

```python
len(courses)
```

```python
codes = [x['course_code'] for x in courses]
codes
```

```python
years = [21,22,23]
```

```python
terms=['t1','t2','t3']
```

```python
namespaces = []
for year in years:
    for term in terms:
        for code in codes:
            namespaces.append(f'ns_{year}{term}_{code[2:].lower()}')
namespaces
```

```python
len(namespaces)
```

```python
with open('namespaces.txt','w') as newfile:
    for ns in namespaces:
        newfile.write(ns+'\n')
```
